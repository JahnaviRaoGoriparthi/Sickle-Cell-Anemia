```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split  # Import random_split
```

Import Statements


```python

def load_dataset(dataset_path, batch_size=32, train_split=0.8):
    """
    Loads dataset from the given path, applies transformations, splits into train/test, and returns DataLoaders.
    """
    # Define transformations
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
    ])

    # Load dataset
    dataset = datasets.ImageFolder(root=dataset_path, transform=transform)

    # Split dataset
    train_size = int(train_split * len(dataset))
    test_size = len(dataset) - train_size
    train_dataset, test_dataset = random_split(dataset, [train_size, test_size])  # Now works fine

    # Create DataLoaders
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

    return train_loader, test_loader, dataset.classes

# Usage
dataset_path = "/Users/yanthraa/Desktop/PROJECT/CODE/BreaKHis 400X/data"
train_loader, test_loader, class_names = load_dataset(dataset_path)

print("Classes:", class_names)
print("Benign samples:", len(train_loader.dataset))
print(" samples:", len(test_loader.dataset))
```

Mean & Standard Deviation Calculation for Normalization

This will give you the mean and standard deviation for each color channel (R, G, B). You can use these values to normalize images when training your model.

```python
import torch
import numpy as np

# Calculate mean and std for the dataset
def calculate_mean_std(data_loader):
    mean = 0.
    std = 0.
    num_samples = 0

    for images, _ in data_loader:
        batch_samples = images.size(0)  # Batch size
        images = images.view(batch_samples, images.size(1), -1)  # Flatten image
        mean += images.mean(dim=2).mean(dim=0)  # Compute mean per channel
        std += images.std(dim=2).std(dim=0)  # Compute std per channel
        num_samples += batch_samples

    mean /= num_samples
    std /= num_samples

    return mean, std

# Calculate for training dataset
mean, std = calculate_mean_std(train_loader)

print("Mean:", mean)
print("Std:", std)
```

PCA (Principal Component Analysis):

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Flatten the images for PCA
def get_flattened_images(data_loader):
    flattened_images = []
    labels = []
    for image, label in data_loader:
        flattened_images.append(image.view(image.size(0), -1).numpy())  # Flatten each image
        labels.extend(label.numpy())  # Append labels
    return np.concatenate(flattened_images), np.array(labels)

# Get flattened images and their labels
flattened_images, labels = get_flattened_images(train_loader)

# Apply PCA for dimensionality reduction (to 2D)
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(flattened_images)

# Plot results
plt.figure(figsize=(8, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap='viridis', s=10)
plt.title("PCA of Dataset (2D Projection)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.colorbar()
plt.show()
```

t-SNE (t-distributed Stochastic Neighbor Embedding):

Feature Maps Visualization (for CNN models)

Once you build a Convolutional Neural Network (CNN), you can visualize the feature maps learned by the network during training. Below is an example of how to visualize feature maps after passing an image through a CNN.

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Example CNN model for visualization
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 64, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.fc1 = nn.Linear(128*224*224, 10)  # Dummy output for 10 classes

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = x.view(x.size(0), -1)  # Flatten
        x = self.fc1(x)
        return x

# Load an image and pass it through the network
def visualize_feature_maps(model, image):
    model.eval()  # Set model to evaluation mode

    # Hook to capture feature maps
    def hook_fn(module, input, output):
        feature_maps.append(output)

    feature_maps = []  # List to store feature maps
    hook_handle = model.conv2.register_forward_hook(hook_fn)  # Attach hook to conv2 layer

    with torch.no_grad():
        model(image.unsqueeze(0))  # Add batch dimension and forward pass

    hook_handle.remove()  # Remove the hook after forward pass

    # Visualize the feature maps
    num_features = feature_maps[0].shape[1]  # Number of feature maps
    plt.figure(figsize=(12, 12))
    for i in range(min(num_features, 64)):  # Show up to 64 feature maps
        plt.subplot(8, 8, i+1)
        plt.imshow(feature_maps[0][0, i].detach().numpy(), cmap='gray')
        plt.axis('off')
    plt.show()

# Visualize feature maps for a sample image
image, _ = train_loader.dataset[0]  # Load a sample image
visualize_feature_maps(SimpleCNN(), image)
```

```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Subset
import numpy as np
from collections import Counter

def load_dataset(dataset_path, batch_size=32, train_split=0.8):
    """
    Loads dataset from the given path, applies transformations, splits into train/test, 
    and returns DataLoaders. Ensures proportional class distribution in both splits.
    """
    # Define transformations (resize and tensor conversion)
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
    ])

    # Load dataset using ImageFolder
    dataset = datasets.ImageFolder(root=dataset_path, transform=transform)

    # Create class-specific indices (class 0 = benign, class 1 = malignant)
    class_indices = {0: [], 1: []}
    
    # Store the indices for each class
    for idx, (_, label) in enumerate(dataset.samples):
        class_indices[label].append(idx)

    # Split each class independently
    train_indices = []
    test_indices = []
    
    for label, indices in class_indices.items():
        # Calculate the split
        split = int(train_split * len(indices))
        
        # Shuffle indices for randomness
        np.random.shuffle(indices)
        
        # Add to train and test indices
        train_indices.extend(indices[:split])
        test_indices.extend(indices[split:])
    
    # Create subsets for train and test using the indices
    train_dataset = Subset(dataset, train_indices)
    test_dataset = Subset(dataset, test_indices)

    # Create DataLoaders
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

    return train_loader, test_loader, dataset.classes, len(train_indices), len(test_indices)

# Usage
dataset_path = "/Users/yanthraa/Desktop/PROJECT/CODE/BreaKHis 400X/data"
train_loader, test_loader, class_names, train_size, test_size = load_dataset(dataset_path)

# Check the class distribution
train_labels = [label for _, label in train_loader.dataset]
test_labels = [label for _, label in test_loader.dataset]

train_class_distribution = dict(Counter(train_labels))
test_class_distribution = dict(Counter(test_labels))

# Print results
print("Classes:", class_names)
print(f"Total samples: {train_size + test_size}")
print(f"Training samples: {train_size}, Testing samples: {test_size}")
print(f"Class distribution in training set: {train_class_distribution}")
print(f"Class distribution in testing set: {test_class_distribution}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to show a batch of augmented images
def show_augmented_images(loader):
    data_iter = iter(loader)
    images, labels = next(data_iter)
    
    # Convert images from tensor to numpy for visualization
    images = images.numpy().transpose(0, 2, 3, 1)  # Convert to (batch, height, width, channels)
    
    # Clip values to [0, 1] for display, if needed (to handle any out-of-range values)
    images_clipped = np.clip(images, 0, 1)

    # Plot a few images from the batch
    fig, axes = plt.subplots(1, 5, figsize=(15, 5))
    for i in range(5):
        axes[i].imshow(images_clipped[i])
        axes[i].axis('off')
    plt.show()

# Visualize augmented images from the training data
show_augmented_images(train_loader)
```
