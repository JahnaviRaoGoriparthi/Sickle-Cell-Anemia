```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Define constants
data_path = "/Users/yanthraa/Desktop/PROJECT/NEW DATASET/archive"
target_size = (224, 224)  # Define the target size for resizing

# Load and preprocess data
images = []
labels = []

# Loop through each class (Negatives and Positives)
for class_name in ['Negatives', 'Positives']:
    class_path = os.path.join(data_path, class_name)
    # Iterate through each image in the class
    for image_name in os.listdir(class_path):
        image_path = os.path.join(class_path, image_name)
        # Read the image in grayscale
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        # Resize the image to the target size
        img = cv2.resize(img, target_size)
        # Negativesize the pixel values to the range [0, 1]
        img = img.astype(np.float32) / 255.0
        # Append the preprocessed image to the images list
        images.append(img)
        # Append the label (0 for Negatives, 1 for Positives) to the labels list
        labels.append(0 if class_name == 'Negatives' else 1)

# Convert images and labels to NumPy arrays
images = np.array(images)
labels = np.array(labels)

# Print class proportions
print("Class Proportions:")
print("Negatives:", np.sum(labels == 0), "Positives:", np.sum(labels == 1))

# Plot histogram for class distribution
plt.figure(figsize=(8, 6))
plt.hist(labels, bins=2, color='skyblue', edgecolor='black', alpha=0.7)
plt.xticks([0.25, 0.75], ['Negatives', 'Positives'])
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Data Distribution')
plt.show()
```

Histogram of Image Pixel Intensities


The pixel intensity histogram helps visualize the distribution of pixel values in an image dataset, providing insights into contrast, normalization, and potential outliers. It can guide preprocessing decisions and highlight class distribution differences, which may impact model performance.

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# Function to get image tensors from both classes
def get_image_tensors(dataset, num_samples=5):
    # Separate images by class
    negatives = [dataset[i][0] for i in range(len(dataset)) if dataset[i][1] == 0][:num_samples]
    positives = [dataset[i][0] for i in range(len(dataset)) if dataset[i][1] == 1][:num_samples]

    # Combine the images from both classes
    image_tensors = torch.stack(negatives + positives)
    return image_tensors

# Get image tensors from 5 images from each class
image_tensors = get_image_tensors(train_loader.dataset, num_samples=5)

# Flatten the image tensors into a single array
image_tensors = image_tensors.view(image_tensors.size(0), -1).numpy()

# Plot histogram of pixel intensities
plt.figure(figsize=(8, 5))
plt.hist(image_tensors.ravel(), bins=50, color='purple', alpha=0.7)
plt.title("Histogram of Image Pixel Intensities")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

Mean & Standard Deviation Calculation for Normalization

This will give you the mean and standard deviation for each color channel (R, G, B). You can use these values to normalize images when training your model.

```python
import torch
import numpy as np

# Calculate mean and std for the dataset
def calculate_mean_std(data_loader):
    mean = 0.
    std = 0.
    num_samples = 0

    for images, _ in data_loader:
        batch_samples = images.size(0)  # Batch size
        images = images.view(batch_samples, images.size(1), -1)  # Flatten image dimensions (height, width)
        
        # Compute mean and std for each channel across the batch and image dimensions
        mean += images.mean(dim=(0, 2))  # Mean over height and width
        std += images.std(dim=(0, 2))    # Std over height and width
        num_samples += batch_samples

    mean /= num_samples
    std /= num_samples

    return mean, std

# Calculate for training dataset
mean, std = calculate_mean_std(train_loader)

print("Mean:", mean)
print("Std:", std)
```

PCA (Principal Component Analysis):

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import torch
import numpy as np

# Flatten the images for PCA
def get_flattened_images(data_loader):
    flattened_images = []
    labels = []
    for images, target_labels in data_loader:
        flattened_images.append(images.view(images.size(0), -1).numpy())  # Flatten each image
        labels.extend(target_labels.numpy())  # Append labels
    return np.concatenate(flattened_images), np.array(labels)

# Get flattened images and their labels
flattened_images, labels = get_flattened_images(train_loader)

# Apply PCA for dimensionality reduction (to 2D)
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(flattened_images)

# Plot results
plt.figure(figsize=(8, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap='viridis', s=10)
plt.title("PCA of Dataset (2D Projection)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.colorbar(label='Class Label')
plt.show()
```

t-SNE (t-distributed Stochastic Neighbor Embedding):

t-SNE is better for visualizing clusters in high-dimensional data, especially when trying to visualize class separability.

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Apply t-SNE for dimensionality reduction (to 2D)
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
reduced_data_tsne = tsne.fit_transform(flattened_images)

# Plot t-SNE results
plt.figure(figsize=(8, 6))
plt.scatter(reduced_data_tsne[:, 0], reduced_data_tsne[:, 1], c=labels, cmap='viridis', s=10)
plt.title("t-SNE of Dataset (2D Projection)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.colorbar(label='Class Label')
plt.show()
```
